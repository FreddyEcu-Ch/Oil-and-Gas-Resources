```python
# Function to create various plot from well logs
def plot_curves(start, stop, curves_to_plot):
    fig, ax = plt.subplots(1, len(curves_to_plot), figsize=(20,10))
    for i, log in enumerate(curves_to_plot):
        segment = w_log.data[log].to_basis(start=start, stop=stop)
        segment.plot(ax=ax[i])
        ax[i].set_title(log)
```

```python
curves_to_plot = ['RT', 'SP', 'GR']

plot_curves(3150, 3400, curves_to_plot)
```

```python
curves_to_plot = ['RT', 'SP', 'GR', 'NPHI', 'RHOB', 'SWE_SIM_F3', 'VSH_FINAL_F3']

@interact
def depth_ranges(min_depth= (3000, 3400, 1), max_depth= (3000, 3400, 1)):
    return plot_curves(min_depth, max_depth, curves_to_plot)
```
